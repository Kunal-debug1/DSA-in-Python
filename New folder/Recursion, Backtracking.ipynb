**🧠 PART 1: RECURSION BASICS**

Recursion = Function calling itself.

Every recursive problem has:

1️⃣ Base case

2️⃣ Recursive call

3️⃣ Smaller subproblem

# 🟢 BASIC RECURSION PROBLEMS

**1️⃣ Factorial**

```python
def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    
    return n * factorial(n - 1)  # Recursive call


print(factorial(5))
```

**2️⃣ Fibonacci**

```python
def fibonacci(n, memo={}):
    # Step 1: If already computed, return stored value
    if n in memo:
        return memo[n]

    # Step 2: Base cases
    if n <= 1:
        return n

    # Step 3: Compute and store result in memo
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)

    return memo[n]


# -------------------------
# Example Input
# -------------------------

n = 6
print("Fibonacci of", n, "is:", fibonacci(n))
```

**3️⃣ Power (x^n)**

```python
def power(x, n):
    # Base case
    if n == 0:
        return 1
    
    # Recursive multiplication
    return x * power(x, n-1)


print(power(2, 5))
```

# 🟡 PART 2: BACKTRACKING BASICS

**Backtracking = Try → Explore → Undo → Try next**

Used in:

- Subsets
- Combinations
- Permutations
- Parentheses generation

# 🥇 IMPORTANT BACKTRACKING PROBLEMS

1️⃣ Subsets (VERY IMPORTANT)

Input:

nums = [1,2,3]

```python
def subsets(nums):
    result = []  # Stores all subsets

    def backtrack(start, path):
        # Add current subset (make copy!)
        result.append(path[:])

        # Try adding each remaining number
        for i in range(start, len(nums)):
            path.append(nums[i])        # Choose
            backtrack(i + 1, path)      # Explore
            path.pop()                  # Un-choose (backtrack)

    backtrack(0, [])
    return result


# Example
print(subsets([1,2,3]))
```

**2️⃣ Permutations**

```python
def permutations(nums):
    result = []  # Stores all permutations

    def backtrack(path):
        # Base case: if permutation is complete
        if len(path) == len(nums):
            result.append(path[:])  # Make copy
            return

        for num in nums:
            # Skip if number already used in current permutation
            if num not in path:
                path.append(num)       # Choose
                backtrack(path)        # Explore
                path.pop()             # Undo (backtrack)

    backtrack([])
    return result


print(permutations([1,2,3]))
```

**3️⃣ Combination Sum**

```python
def combination_sum(candidates, target):
    result = []

    def backtrack(start, path, total):
        # ✅ If we reach exact target → store combination
        if total == target:
            result.append(path[:])
            return
        
        # ❌ If total exceeds target → stop exploring
        if total > target:
            return

        # Try all candidates starting from current index
        for i in range(start, len(candidates)):
            path.append(candidates[i])   # Choose
            backtrack(i, path, total + candidates[i])  # Reuse allowed (i, not i+1)
            path.pop()   # Undo choice (backtrack)

    backtrack(0, [], 0)
    return result


print(combination_sum([2,3,6,7], 7))
```

**4️⃣ Generate Parentheses**

```python
def generate_parentheses(n):
    result = []

    def backtrack(open_count, close_count, path):
        # If the string length reaches 2*n, it's valid
        if len(path) == 2 * n:
            result.append(path)
            return

        # Add "(" if we still have openings left
        if open_count < n:
            backtrack(open_count + 1, close_count, path + "(")

        # Add ")" only if it won't make invalid string
        if close_count < open_count:
            backtrack(open_count, close_count + 1, path + ")")

    backtrack(0, 0, "")
    return result


print(generate_parentheses(3))
```
