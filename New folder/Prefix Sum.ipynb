# 🧠 WHAT IS PREFIX SUM?

Prefix sum means:

Instead of recalculating sum again and again,
we store cumulative sum.

Example:

Array: [1,2,3,4]

Prefix: [1,3,6,10]

Now sum of subarray from index i to j:

sum(i,j) = prefix[j] - prefix[i-1]

This makes operations O(1).

**1️⃣ Range Sum Query (Basic Prefix)**

📝 Input:

arr = [1,2,3,4,5]

Query: sum(1,3) → 2+3+4 = 9

```python
def build_prefix(arr):
    # Create prefix array of same length
    prefix = [0] * len(arr)
    
    # First element stays same
    prefix[0] = arr[0]

    # Build prefix sum
    for i in range(1, len(arr)):
        prefix[i] = prefix[i-1] + arr[i]

    return prefix


def range_sum(prefix, left, right):
    # If range starts from 0
    if left == 0:
        return prefix[right]

    # Otherwise subtract previous prefix
    return prefix[right] - prefix[left-1]


# Example
arr = [1, 2, 3, 4, 5]
prefix = build_prefix(arr)

print(range_sum(prefix, 1, 3))
```

**2️⃣ Subarray Sum Equals K (VERY IMPORTANT)**

📝 Input:

nums = [1,1,1], k = 2

```python
def subarray_sum(nums, k):
    prefix_sum = 0          # Running sum
    count = 0               # Number of valid subarrays
    hashmap = {0: 1}        # {prefix_sum: frequency}
                            # Important for subarrays starting at index 0

    for num in nums:
        # Step 1: Update running prefix sum
        prefix_sum += num

        # Step 2: Check if there exists a prefix such that:
        # prefix_sum - previous_prefix = k
        # => previous_prefix = prefix_sum - k
        if prefix_sum - k in hashmap:
            count += hashmap[prefix_sum - k]

        # Step 3: Store current prefix sum
        hashmap[prefix_sum] = hashmap.get(prefix_sum, 0) + 1

    return count


# Example
print(subarray_sum([1,1,1], 2))
```

**3️⃣ Longest Subarray with Equal 0s and 1s**

📝 Idea:

Convert 0 → -1

Then find subarray sum = 0

```python
def find_max_length(nums):
    prefix = 0
    hashmap = {0: -1}  # Stores first index of each prefix sum
                       # {prefix_sum : first_index}
    max_len = 0

    for i in range(len(nums)):

        # Treat 0 as -1
        if nums[i] == 0:
            prefix -= 1
        else:
            prefix += 1

        # If prefix seen before → subarray sum = 0
        if prefix in hashmap:
            max_len = max(max_len, i - hashmap[prefix])
        else:
            # Store first occurrence only
            hashmap[prefix] = i

    return max_len


print(find_max_length([0,1,0]))
```

**4️⃣ Subarray Sums Divisible by K**

📝Input:

nums = [4,5,0,-2,-3,1], k = 5

```python
def subarrays_div_by_k(nums, k):
    prefix = 0
    count = 0
    hashmap = {0: 1}  # {remainder : frequency}

    for num in nums:
        # Step 1: Update prefix sum
        prefix += num
        
        # Step 2: Compute remainder
        remainder = prefix % k
        
        # ⚠ Important for negative numbers
        # In some languages you'd write:
        # remainder = (prefix % k + k) % k

        # Step 3: If remainder seen before → valid subarray(s)
        if remainder in hashmap:
            count += hashmap[remainder]

        # Step 4: Store remainder
        hashmap[remainder] = hashmap.get(remainder, 0) + 1

    return count


print(subarrays_div_by_k([4,5,0,-2,-3,1], 5))
```

**5️⃣ Count Nice Subarrays (Exactly K Odds)**

```python
def number_of_subarrays(nums, k):
    prefix = {0: 1}   # {odd_count: frequency}
                      # Important for subarrays starting from index 0
    
    odd_count = 0     # Running count of odd numbers
    count = 0         # Number of valid subarrays

    for num in nums:

        # Step 1: Increase odd count if number is odd
        if num % 2 == 1:
            odd_count += 1

        # Step 2: Check if we previously had (odd_count - k)
        # If yes → subarray with exactly k odds exists
        if odd_count - k in prefix:
            count += prefix[odd_count - k]

        # Step 3: Store current odd_count
        prefix[odd_count] = prefix.get(odd_count, 0) + 1

    return count
print(number_of_subarrays([1,1,2,1,1], 3))
```

**6️⃣ Product of Array Except Self (Prefix + Suffix)**

```python
def product_except_self(nums):
    n = len(nums)
    
    left = [1] * n    # left[i] = product of all elements to the left of i
    right = [1] * n   # right[i] = product of all elements to the right of i
    result = [1] * n

    # Step 1: Build left product array
    for i in range(1, n):
        left[i] = left[i-1] * nums[i-1]

    # Step 2: Build right product array
    for i in range(n-2, -1, -1):
        right[i] = right[i+1] * nums[i+1]

    # Step 3: Multiply left and right
    for i in range(n):
        result[i] = left[i] * right[i]

    return result


print(product_except_self([1,2,3,4]))
```
