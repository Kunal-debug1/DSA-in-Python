# 🟢 1️⃣ BASIC BINARY SEARCH

**1️⃣ Classic Binary Search**

📝 Input:

arr = [1,3,5,7,9], target = 7

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2   # Middle index

        if arr[mid] == target:
            return mid              # Target found
        
        elif arr[mid] < target:
            left = mid + 1          # Search right half
        
        else:
            right = mid - 1         # Search left half

    return -1   # Target not found


print(binary_search([1,3,5,7,9], 7))
```

# 🟡 2️⃣ LOWER BOUND / FIRST OCCURRENCE
**2️⃣ Find First Occurrence**

```python
def first_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1   # Stores first occurrence index

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid
            right = mid - 1   # Keep searching LEFT side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result

print(first_occurrence([1,2,2,2,3,4], 2))
```

**3️⃣ Find Last Occurrence**

```python
def last_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1   # Stores last occurrence index

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid
            left = mid + 1   # Keep searching RIGHT side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result

print(last_occurrence([1,2,2,2,3,4], 2))
```

# 🟡 3️⃣ SEARCH IN ROTATED SORTED ARRAY
**4️⃣ Search in Rotated Sorted Array**

```python
def search_rotated(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        # If found
        if nums[mid] == target:
            return mid

        # Case 1: Left half is sorted
        if nums[left] <= nums[mid]:

            # Target lies inside sorted left half
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1

        # Case 2: Right half is sorted
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1

print(search_rotated([4,5,6,7,0,1,2], 0))
```

# 🟡 4️⃣ FIND PEAK ELEMENT
**5️⃣ Peak Index in Mountain Array**

```python
def peak_index(arr):
    left, right = 0, len(arr) - 1

    while left < right:
        mid = (left + right) // 2

        # If middle is smaller than next element
        # Peak must be on the right side
        if arr[mid] < arr[mid + 1]:
            left = mid + 1
        else:
            # Peak is on the left side (including mid)
            right = mid

    # left == right → peak index
    return left

print(peak_index([1,2,3,1]))
```

# 🔥 5️⃣ BINARY SEARCH ON ANSWER (VERY IMPORTANT)

This is advanced but very powerful.

**6️⃣ Sqrt(x)**

```python
def my_sqrt(x):
    # Edge case: sqrt(0) = 0, sqrt(1) = 1
    if x < 2:
        return x

    left, right = 0, x
    answer = 0  # Store best valid square root found so far

    while left <= right:
        mid = (left + right) // 2

        # If mid^2 is less than or equal to x,
        # it is a possible answer
        if mid * mid <= x:
            answer = mid       # Store current valid value
            left = mid + 1     # Try to find larger valid value
        else:
            right = mid - 1    # Mid is too large

    return answer


# -------------------------
# Example Input
# -------------------------

x = 8
print("Square root of", x, "is:", my_sqrt(x))
```

**7️⃣ Minimum Eating Speed (Koko Bananas Type)**

Search minimum possible speed.

```python
def min_eating_speed(piles, h):
    import math

    # Minimum possible speed = 1
    # Maximum possible speed = max pile
    left, right = 1, max(piles)

    while left < right:
        mid = (left + right) // 2   # Candidate eating speed

        # Calculate total hours needed at speed = mid
        hours = sum(math.ceil(p / mid) for p in piles)

        # If she can finish in h hours or less,
        # try smaller speed
        if hours <= h:
            right = mid
        else:
            # Speed too slow → increase speed
            left = mid + 1

    return left


# -------------------------
# Example Input
# -------------------------

piles = [3,6,7,11]
h = 8

print("Minimum eating speed:", min_eating_speed(piles, h))
```
