# 🥇 MOST IMPORTANT GREEDY PROBLEMS (With Optimized Solutions)

**1️⃣ Best Time to Buy and Sell Stock (Single Transaction)**

📝 Input:

prices = [7,1,5,3,6,4]

🧠 Idea:

*Track minimum price so far.*


```python
def max_profit(prices):
    min_price = float('inf')   # Track minimum price seen so far
    max_profit = 0             # Track maximum profit

    for price in prices:
        # Update minimum buying price
        min_price = min(min_price, price)

        # Calculate profit if sold today
        profit = price - min_price

        # Update maximum profit
        max_profit = max(max_profit, profit)

    return max_profit


print(max_profit([7,1,5,3,6,4]))
```

⏱ O(n)

🔥 HIGH PRIORITY FOR TCS

**2️⃣ Best Time to Buy and Sell Stock II (Multiple Transactions)**

```python
def max_profit2(prices):
    profit = 0  # Total accumulated profit

    for i in range(1, len(prices)):

        # If price increased from yesterday
        if prices[i] > prices[i - 1]:
            
            # Take that profit
            profit += prices[i] - prices[i - 1]

    return profit

print(max_profit2([7,1,5,3,6,4]))
```

**3️⃣ Jump Game**

🧠 Idea:

Track maximum reachable index.

```python
def can_jump(nums):
    max_reach = 0  # Furthest index we can reach so far

    for i in range(len(nums)):
        
        # If current index is beyond what we can reach
        if i > max_reach:
            return False

        # Update the furthest reachable index
        max_reach = max(max_reach, i + nums[i])

    return True


print(can_jump([2,3,1,1,4]))
```

**4️⃣ Jump Game II (Minimum Jumps)**

```python
def jump(nums):
    jumps = 0          # Number of jumps made
    current_end = 0    # End of current jump range
    farthest = 0       # Farthest index reachable in current range

    # We stop at len(nums)-1 because we don't need to jump from last index
    for i in range(len(nums) - 1):

        # Update farthest reachable index
        farthest = max(farthest, i + nums[i])

        # If we reach the end of current jump range
        if i == current_end:
            jumps += 1              # We must jump
            current_end = farthest  # Update range for next jump

    return jumps

print(jump([2,3,1,1,4]))
```

**5️⃣ Maximum Subarray (Kadane’s Algorithm)**

```python
def max_subarray(nums):
    # Start with first element
    current = nums[0]   # Current subarray sum
    maximum = nums[0]   # Maximum subarray sum so far

    for i in range(1, len(nums)):
        # Either start new subarray from nums[i]
        # OR extend previous subarray
        current = max(nums[i], current + nums[i])

        # Update global maximum
        maximum = max(maximum, current)

    return maximum

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

**6️⃣ Maximum Product Subarray**

```python
def max_product(nums):
    current_max = nums[0]
    current_min = nums[0]
    result = nums[0]

    for i in range(1, len(nums)):
        num = nums[i]

        # If number is negative, swap max & min
        if num < 0:
            current_max, current_min = current_min, current_max

        # Either start fresh from num
        # OR multiply with previous product
        current_max = max(num, current_max * num)
        current_min = min(num, current_min * num)

        # Update global maximum
        result = max(result, current_max)

    return result


print(max_product([2,3,-2,4]))
```

**7️⃣ Assign Cookies**

🧠 Match smallest greed first

```python
def find_content_children(g, s):
    # Step 1: Sort greed factors and cookie sizes
    g.sort()
    s.sort()

    i = j = 0  # i → child pointer, j → cookie pointer

    # Step 2: Try to assign cookies
    while i < len(g) and j < len(s):
        
        # If current cookie satisfies child
        if s[j] >= g[i]:
            i += 1   # Child satisfied
        
        # Move to next cookie
        j += 1

    # i represents number of satisfied children
    return i

print(find_content_children([1,2,3], [1,1]))
```

**8️⃣ Gas Station**

```python
def can_complete_circuit(gas, cost):
    total = 0   # Total net gas across all stations
    tank = 0    # Current gas tank
    start = 0   # Candidate starting station

    for i in range(len(gas)):
        diff = gas[i] - cost[i]

        total += diff   # Overall feasibility check
        tank += diff    # Current running tank

        # If tank becomes negative,
        # we cannot start from previous start
        if tank < 0:
            start = i + 1
            tank = 0

    # If total gas is negative, impossible
    return start if total >= 0 else -1

gas  = [1,2,3,4,5]
cost = [3,4,5,1,2]

print(can_complete_circuit(gas, cost))
```

**9️⃣ Partition Labels**

```python
def partition_labels(s):
    # Step 1: Store last occurrence of each character
    last = {c: i for i, c in enumerate(s)}

    size = 0      # Current partition size
    end = 0       # End boundary of current partition
    result = []   # Final partition sizes

    # Step 2: Traverse string
    for i, c in enumerate(s):

        # Update partition end to furthest last occurrence
        end = max(end, last[c])

        size += 1

        # If we reach the partition boundary
        if i == end:
            result.append(size)
            size = 0

    return result

print(partition_labels("ababcbacadefegdehijhklij"))
```

**🔟 Candy Distribution**

```python
def candy(ratings):
    n = len(ratings)
    
    # Step 1: Give every child at least 1 candy
    candies = [1] * n

    # Step 2: Left → Right pass
    # If rating increases, increase candy count
    for i in range(1, n):
        if ratings[i] > ratings[i - 1]:
            candies[i] = candies[i - 1] + 1

    # Step 3: Right → Left pass
    # Ensure right neighbor condition
    for i in range(n - 2, -1, -1):
        if ratings[i] > ratings[i + 1]:
            candies[i] = max(candies[i], candies[i + 1] + 1)

    # Step 4: Total candies
    return sum(candies)

print(candy([1,0,2]))
```
