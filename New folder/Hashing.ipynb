# 🥇 HASHING / FREQUENCY PATTERN – MOST IMPORTANT QUESTIONS
**1️⃣ Two Sum (Unsorted Array)**

🧠 Idea:

*Store numbers in dictionary (value → index)*

📝 Input:

nums = [2,7,11,15], target = 9

✅ Solution:

```python
def two_sum(nums, target):
    hashmap = {}

    for i in range(len(nums)):
        complement = target - nums[i]

        # If complement already seen
        if complement in hashmap:
            return [hashmap[complement], i]

        # Store current number index
        hashmap[nums[i]] = i

# Example
print(two_sum([2,7,11,15], 9))
```

**2️⃣ Majority Element (> n/2)**

📝 Input:

nums = [2,2,1,1,1,2,2]

✅ Solution (HashMap)
⏱ O(n)

```python
def majority_element(nums):
    freq = {}

    for num in nums:
        freq[num] = freq.get(num, 0) + 1

        if freq[num] > len(nums)//2:
            return num

print(majority_element([2,2,1,1,1,2,2]))
```

**3️⃣ Contains Duplicate**

📝 Input:

[1,2,3,1]

```python
def contains_duplicate(nums):
    seen = set()  # Set to store elements we have already seen

    for num in nums:
        # If number already exists → duplicate found
        if num in seen:
            return True
        
        # Otherwise add it to set
        seen.add(num)

    # If loop finishes → no duplicates
    return False


# Example
nums = [1, 2, 3, 1]
print(contains_duplicate(nums))
```

**4️⃣ Valid Anagram**

📝 Input:

s = "anagram", t = "nagaram"

```python
def is_anagram(s, t):
    # If lengths differ → cannot be anagram
    if len(s) != len(t):
        return False

    freq = {}  # Dictionary to store character frequency

    # Step 1: Count characters in string s
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    # Step 2: Subtract frequency using string t
    for ch in t:
        # If character not present or used more times → not anagram
        if ch not in freq or freq[ch] == 0:
            return False
        freq[ch] -= 1

    # If all matched properly
    return True


# Example
print(is_anagram("anagram", "nagaram"))
```

**5️⃣ First Unique Character in String**

📝 Input:

"leetcode"

```python
def first_unique_char(s):
    freq = {}  # Dictionary to store character frequency

    # Step 1: Count frequency of each character
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    # Step 2: Find first character with frequency 1
    for i in range(len(s)):
        if freq[s[i]] == 1:
            return i

    # If no unique character found
    return -1


# Example
print(first_unique_char("leetcode"))
```

**6️⃣ Subarray Sum Equals K (VERY IMPORTANT)**

📝 Input:

nums = [1,1,1], k = 2

```python
def subarray_sum(nums, k):
    prefix_sum = 0          # Running sum
    count = 0               # Number of subarrays with sum = k
    hashmap = {0: 1}        # Stores frequency of prefix sums
                             # {prefix_sum: occurrences}

    for num in nums:
        # Step 1: Update running prefix sum
        prefix_sum += num

        # Step 2: Check if there exists a prefix such that:
        # prefix_sum - previous_prefix = k
        # => previous_prefix = prefix_sum - k
        if prefix_sum - k in hashmap:
            count += hashmap[prefix_sum - k]

        # Step 3: Store current prefix sum in hashmap
        hashmap[prefix_sum] = hashmap.get(prefix_sum, 0) + 1

    return count


# Example
print(subarray_sum([1,1,1], 2))
```

**7️⃣ Longest Consecutive Sequence**

📝 Input:

[100,4,200,1,3,2]

```python
def longest_consecutive(nums):
    # Convert list to set for O(1) lookups
    num_set = set(nums)
    longest = 0

    for num in num_set:

        # Only start counting if 'num' is the beginning of a sequence
        # i.e., previous number does NOT exist
        if num - 1 not in num_set:
            
            length = 1
            
            # Keep checking next consecutive numbers
            while num + length in num_set:
                length += 1

            # Update longest sequence found
            longest = max(longest, length)

    return longest


# Example
print(longest_consecutive([100,4,200,1,3,2]))
```

**8️⃣ K-diff Pairs in Array**

📝 Input:

nums = [3,1,4,1,5], k = 2

```python
def find_pairs(nums, k):
    # Step 1: Edge case
    if k < 0:
        return 0   # Difference cannot be negative

    freq = {}      # Frequency dictionary
    count = 0      # Number of unique k-diff pairs

    # Step 2: Count frequency of each number
    for num in nums:
        freq[num] = freq.get(num, 0) + 1

    # Step 3: Check pairs
    for num in freq:

        # Case 1: k == 0
        # We need duplicates
        if k == 0:
            if freq[num] > 1:
                count += 1

        # Case 2: k > 0
        # Check if num + k exists
        else:
            if num + k in freq:
                count += 1

    return count


# Example
print(find_pairs([3,1,4,1,5], 2))
```

**9️⃣ Count Nice Subarrays (Exactly K odd numbers)**

📝 Input:

nums = [1,1,2,1,1], k = 3

```python
def number_of_subarrays(nums, k):
    count = 0
    prefix = {0: 1}   # Stores frequency of odd counts
    odd_count = 0     # Running count of odd numbers

    for num in nums:

        # Step 1: Increase odd count if number is odd
        if num % 2 == 1:
            odd_count += 1

        # Step 2: Check if we previously had (odd_count - k)
        # If yes → subarray with exactly k odds exists
        if odd_count - k in prefix:
            count += prefix[odd_count - k]

        # Step 3: Store current odd_count
        prefix[odd_count] = prefix.get(odd_count, 0) + 1

    return count


# Example
print(number_of_subarrays([1,1,2,1,1], 3))
```

🔟 Group Anagrams
📝 Input:

["eat","tea","tan","ate","nat","bat"]

```python
def group_anagrams(strs):
    from collections import defaultdict
    
    hashmap = defaultdict(list)  # {sorted_word : list_of_anagrams}

    for word in strs:
        # Sort characters to create a unique key
        # Anagrams will have the same sorted version
        key = "".join(sorted(word))
        
        # Append original word under this key
        hashmap[key].append(word)

    # Return grouped anagrams
    return list(hashmap.values())


# Example
print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```
