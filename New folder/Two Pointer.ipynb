# 🔥 WHAT IS TWO POINTER?

Two common types:

1️⃣ Opposite Direction → left & right

2️⃣ Same Direction (Slow/Fast) → remove duplicates, shifting elements

🟢 EASY LEVEL

**🧠 Two Pointer Recognition Trick**

If problem says:

- Sorted array
- Remove duplicates
- Pair sum
- Reverse
- Max/min from ends

👉 Think Two Pointer immediately.

**1️⃣ Two Sum (Sorted Version)**

🧠 Idea:

- If array sorted →
- If sum too small → move left++
- If sum too large → move right--

```python
def two_sum_sorted(arr, target):
    left, right = 0, len(arr)-1

    while left < right:
        s = arr[left] + arr[right]
        if s == target:
            return [left, right]
        elif s < target:
            left += 1
        else:
            right -= 1
    return []

arr = [1,2,3,4,5,6]
target = 9
two_sum_sorted(arr, target)
print(two_sum_sorted(arr, target))
```

**2️⃣ Move Zeroes**

🧠 Idea:

i pointer places non-zero elements.

```python
def move_zeroes(arr):
    i = 0
    for j in range(len(arr)):
        if arr[j] != 0:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    return arr

arr = [1,0,2,0,3,4,0,5,6]
move_zeroes(arr)
print(move_zeroes(arr))
```

**3️⃣ Remove Duplicates from Sorted Array**

🧠 Idea:

Keep unique values using slow pointer.

```python
def remove_duplicates(arr):
    if not arr:
        return 0

    i = 0  # points to last unique element
    
    for j in range(1, len(arr)):
        if arr[j] != arr[i]:   # compare with last unique
            i += 1
            arr[i] = arr[j]

    return i + 1


arr = [1,1,2,3,3,4]   # must be sorted
result = remove_duplicates(arr)

print("New Length:", result)
print("Modified Array:", arr[:result])
```

**4️⃣ Reverse String**

```python
def reverse_string(s):
    s = list(s)
    left, right = 0, len(s)-1
    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
    return "".join(s)

s = "Kunal"
reverse_string(s)
print(reverse_string(s))
```

**5️⃣ Merge Sorted Array**

```python
def merge_sorted(nums1, nums2):
    i = j = 0
    result = []
    while i < len(nums1) and j < len(nums2):
        if nums1[i] < nums2[j]:
            result.append(nums1[i])
            i += 1
        else:
            result.append(nums2[j])
            j += 1
    return result + nums1[i:] + nums2[j:]

nums1 = [1,2,3,4,5]
nums2 = [6,7,8,9]
merge_sorted(nums1, nums2)
print(merge_sorted(nums1, nums2))
```

# 🟡 MEDIUM LEVEL

**6️⃣ Container With Most Water**

🧠 Idea:

Move pointer at smaller height.

```python
def max_area(height):
    left, right = 0, len(height)-1
    max_area = 0

    while left < right:
        min_height = min(height[left], height[right])
        area = min_height * (right-left)
        max_area = max(max_area, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_area

height = [1,2,3,4,5,6,7]
max_area(height)
print(max_area(height))
```

**7️⃣ 3Sum**

```python
def three_sum(nums):
    nums.sort()
    result = []

    for i in range(len(nums) - 2):
        
        # Skip duplicate fixed elements
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left, right = i + 1, len(nums) - 1

        while left < right:
            s = nums[i] + nums[left] + nums[right]

            if s == 0:
                result.append([nums[i], nums[left], nums[right]])

                # Skip duplicates for left
                while left < right and nums[left] == nums[left + 1]:
                    left += 1

                # Skip duplicates for right
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

                left += 1
                right -= 1

            elif s < 0:
                left += 1
            else:
                right -= 1

    return result


nums = [-1, 0, 1, 2, -1, -4]
print(three_sum(nums))
```

**8️⃣ 4Sum**

Same logic as 3Sum, just one more loop.

```python
def four_sum(nums, target):
    nums.sort()
    result = []
    n = len(nums)

    for i in range(n - 3):
        
        # Skip duplicate for i
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        for j in range(i + 1, n - 2):
            
            # Skip duplicate for j
            if j > i + 1 and nums[j] == nums[j - 1]:
                continue

            left = j + 1
            right = n - 1

            while left < right:
                s = nums[i] + nums[j] + nums[left] + nums[right]

                if s == target:
                    result.append([nums[i], nums[j], nums[left], nums[right]])

                    # Skip duplicates
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1

                    left += 1
                    right -= 1

                elif s < target:
                    left += 1
                else:
                    right -= 1

    return result


nums = [1, 0, -1, 0, -2, 2]
target = 0
print(four_sum(nums, target))
```

**9️⃣ Sort Colors (Dutch National Flag)**

```python
def sort_colors(nums):
    low = mid = 0
    high = len(nums)-1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
    return nums

nums = [2,0,2,1,1,0]
print(sort_colors(nums))
```

**1️⃣1️⃣ Next Permutation**

```python
def next_permutation(nums):
    i = len(nums)-2

    while i >= 0 and nums[i] >= nums[i+1]:
        i -= 1

    if i >= 0:
        j = len(nums)-1
        while nums[j] <= nums[i]:
            j -= 1
        nums[i], nums[j] = nums[j], nums[i]

    nums[i+1:] = reversed(nums[i+1:])

nums = [5,4,3,2,1]
next_permutation(nums)
print(nums)
```

**1️⃣2️⃣ Trapping Rain Water**

```python
def trap(height):
    left, right = 0, len(height)-1
    left_max = right_max = 0
    water = 0

    while left < right:
        if height[left] < height[right]:
            left_max = max(left_max, height[left])
            water += left_max - height[left]
            left += 1
        else:
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1

    return water


heights = [4,2,0,3,2,5]
print(trap(heights))
```

**1️⃣3️⃣ K-diff Pairs**

```python
def find_pairs(nums, k):
    # Edge case: negative k not allowed
    if k < 0:
        return 0

    nums.sort()  # Sort array to use two-pointer technique
    
    left = 0
    right = 1
    count = 0

    while right < len(nums):

        # Avoid same pointer position
        if left == right:
            right += 1
            continue

        diff = nums[right] - nums[left]

        # Case 1: Found valid pair
        if diff == k:
            count += 1

            # Move left pointer and skip duplicates
            left += 1
            while left < len(nums) and nums[left] == nums[left - 1]:
                left += 1

        # Case 2: Difference too small → increase right
        elif diff < k:
            right += 1

        # Case 3: Difference too large → increase left
        else:
            left += 1

    return count


# -------------------------
# Example Input
# -------------------------

nums = [3, 1, 4, 1, 5]
k = 2

print("Number of unique pairs:", find_pairs(nums, k))
```

# 🟢 EASY LEVEL (Build Foundation)

### 1️⃣ Reverse an Array

Given an array, reverse it using two pointers.

---

```python
def reverse(arr):
    i = 0
    j = len(arr) - 1

    while i < j:
        arr[i], arr[j] = arr[j], arr[i]
        i += 1
        j -= 1
    return arr
    
arr = [1,2,3,4,5]
result = reverse(arr)
print(result)
```

### 2️⃣ Reverse a String

Reverse string without using built-in reverse.

---

```python
def reverseString(s):
    s = list(s)   # convert string to list
    
    i = 0
    j = len(s) - 1

    while i < j:
        s[i], s[j] = s[j], s[i]
        i += 1
        j -= 1
    
    return "".join(s)   # convert list back to string


text = "Python"
result = reverseString(text)
print(result)
```

### 3️⃣ Check Palindrome String

Return True if string is palindrome.

---

```python
def palindrome(text):
    # Convert string to list (optional, string indexing also works)
    text = list(text)
    
    # Initialize two pointers
    i = 0                      # start pointer
    j = len(text) - 1          # end pointer

    # Run loop until pointers meet
    while i < j:
        # If characters don't match → not palindrome
        if text[i] != text[j]:
            return "Not Palindrome"
        
        # Move pointers toward center
        i += 1
        j -= 1

    # If all characters matched
    return "Palindrome"


text = "12121"
result = palindrome(text)
print(result)
```

### 4️⃣ Move All Zeroes to End

Maintain order of non-zero elements.

---

```python
def moveZero(arr):
    i = 0  # position to place next non-zero element
    
    # j scans the array
    for j in range(len(arr)):
        
        # If current element is non-zero
        if arr[j] != 0:
            # Swap with position i
            arr[i], arr[j] = arr[j], arr[i]
            
            # Move i forward
            i += 1
    
    return arr
        

arr = [1,2,0,3,0,4]
result = moveZero(arr)
print(result)
```

### 5️⃣ Remove Duplicates from Sorted Array

Return new length of unique elements.

---

```python
def removeDuplicates(nums):
    # If array is empty
    if not nums:
        return []
    
    i = 0  # place for unique element
    
    # Start from second element
    for j in range(1, len(nums)):
        
        # If new unique element found
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]
    
    # Return only unique portion
    return nums[:i+1]


nums = [1,1,2,2,3,4,4]
result = removeDuplicates(nums)
print(result)
```

### 6️⃣ Merge Two Sorted Arrays

Merge without using sort().

---

```python
##✅ Method 1 — Simple Merge (Using +)
def mergeArr(arr, arr1):
    return arr + arr1


arr = [1,2,3,4]
arr1 = [5,6,7,8]
result = mergeArr(arr, arr1)
print(result)
```

```python
## ✅ Method 2 — Using Loop (Interview Friendly)
def mergeArr(arr, arr1):
    result = []
    
    # Add elements of first array
    for num in arr:
        result.append(num)
    
    # Add elements of second array
    for num in arr1:
        result.append(num)
    
    return result


arr = [1,2,3,4]
arr1 = [5,6,7,8]
result = mergeArr(arr, arr1)
print(result)
```

```python
##✅ Method 3 — Two Pointers (For Sorted Arrays)

def mergeArr(arr, arr1):
    # i pointer for first array
    i = 0
    
    # j pointer for second array
    j = 0
    
    # Result array to store merged elements
    result = []
    
    # Run loop until one array finishes
    while i < len(arr) and j < len(arr1):
        
        # Compare current elements of both arrays
        if arr[i] < arr1[j]:
            # Add smaller element to result
            result.append(arr[i])
            
            # Move pointer of first array
            i += 1
        else:
            # Add smaller element to result
            result.append(arr1[j])
            
            # Move pointer of second array
            j += 1
    
    # If elements remain in first array, add them
    while i < len(arr):
        result.append(arr[i])
        i += 1
        
    # If elements remain in second array, add them
    while j < len(arr1):
        result.append(arr1[j])
        j += 1
    
    # Return merged sorted array
    return result


# Two sorted arrays
arr = [1,3,5,7]
arr1 = [2,4,6,8]

# Function call
print(mergeArr(arr, arr1))
```

### 7️⃣ Valid Palindrome (Ignore non-alphanumeric)

Example: `"A man, a plan, a canal: Panama"` → True

---

```python
def isPalindrome(s):
    i = 0                      # left pointer
    j = len(s) - 1             # right pointer
    
    while i < j:
        
        # Skip non-alphanumeric from left
        while i < j and not s[i].isalnum():
            i += 1
        
        # Skip non-alphanumeric from right
        while i < j and not s[j].isalnum():
            j -= 1
        
        # Compare characters (convert to lowercase)
        if s[i].lower() != s[j].lower():
            return False
        
        # Move pointers inward
        i += 1
        j -= 1
    
    return True


text = "A man, a plan, a canal: Panama"
print(isPalindrome(text))
```

# 🟡 MEDIUM LEVEL (Core Two Pointer Logic)

### 8️⃣ Two Sum (Sorted Array Version)

Return indices of numbers whose sum = target.

---


```python
def two_sum_sorted(arr, target):
    i = 0
    j = len(arr) - 1

    while i < j:
        current_sum = arr[i] + arr[j]
        
        if current_sum == target:
            return i, j          # return indices
        elif current_sum < target:
            i += 1               # need bigger sum
        else:
            j -= 1               # need smaller sum
    
    return None                  # if no pair found


arr = [1,2,3,4,5,6]
target = 9

result = two_sum_sorted(arr, target)
print(result)
```

### 9️⃣ Container With Most Water

Find maximum area between two lines.

---

```python
def maxArea(height):
    left = 0
    right = len(height) - 1
    max_water = 0

    while left < right:
        width = right - left
        h = min(height[left], height[right])
        water = width * h
        max_water = max(max_water, water)

        # Move pointer of smaller height
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water


height = [1,8,6,2,5,4,8,3,7]
print(maxArea(height))
```

### 1️⃣2️⃣ Squares of Sorted Array

Given sorted array with negatives, return sorted squares.

---

```python
def sortedSquares(nums):
    n = len(nums)
    result = [0] * n
    
    left = 0
    right = n - 1
    pos = n - 1
    
    while left <= right:
        left_sq = nums[left] ** 2
        right_sq = nums[right] ** 2
        
        if left_sq > right_sq:
            result[pos] = left_sq
            left += 1
        else:
            result[pos] = right_sq
            right -= 1
        
        pos -= 1
    
    return result


# Example
nums = [-7, -3, 2, 3, 11]
print(sortedSquares(nums))
```
