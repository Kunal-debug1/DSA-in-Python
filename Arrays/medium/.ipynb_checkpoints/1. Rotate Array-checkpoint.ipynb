# 1. Rotate Array 

Given an integer array nums, rotate the array to the right by k steps, where k is non-negative.


**🔥 1️⃣ Optimal Method (Reversal Algorithm) – O(n), O(1)**

*This is interviewer’s favorite 💎*

✅ RIGHT ROTATION
Steps:

- Reverse whole array
- Reverse first k elements
- Reverse remaining n-k elements

```python
def rotateRight(nums, k):
    n = len(nums)
    k = k % n
    
    # Reverse whole array
    i, j = 0, n - 1
    while i < j:
        nums[i], nums[j] = nums[j], nums[i]
        i += 1
        j -= 1
    
    # Reverse first k elements
    i, j = 0, k - 1
    while i < j:
        nums[i], nums[j] = nums[j], nums[i]
        i += 1
        j -= 1
    
    # Reverse remaining elements
    i, j = k, n - 1
    while i < j:
        nums[i], nums[j] = nums[j], nums[i]
        i += 1
        j -= 1


# Driver Code
nums = [1,2,3,4,5,6,7]
k = 3
rotateRight(nums, k)
print(nums)
```

**🔎 Example**

[1,2,3,4,5]

Reverse whole:

[5,4,3,2,1]

Reverse first 2:

[4,5,3,2,1]

Reverse rest:

[4,5,1,2,3]

✔ Done

**✅ LEFT ROTATION**

Steps:

- Reverse first k elements

- Reverse remaining elements

- Reverse whole array

```python
def rotateLeft(arr, d):
    n = len(arr)
    d = d % n

    # helper reverse function
    def reverse(l, r):
        while l < r:
            arr[l], arr[r] = arr[r], arr[l]
            l += 1
            r -= 1

    # Step 1
    reverse(0, d - 1)

    # Step 2
    reverse(d, n - 1)

    # Step 3
    reverse(0, n - 1)


# Driver Code
nums = [1,2,3,4,5,6,7]
d = 3
rotateLeft(nums, d)
print(nums)
```

🔎 Dry Run

Original:

[1,2,3,4,5,6,7]

Reverse first 3:

[3,2,1,4,5,6,7]

Reverse remaining:

[3,2,1,7,6,5,4]

Reverse whole:

[4,5,6,7,1,2,3]

✔ Correct


*⏱ Complexity*

**Time** → O(n)

**Space** → O(1)

Interview best ✔

**🔥 2️⃣ Simple Python Method (Not Interview Preferred)**
- Right Rotate

nums[:] = nums[-k:] + nums[:-k]

- Left Rotate

nums[:] = nums[k:] + nums[:k]

Uses extra space ❌

**🔥 3️⃣ Brute Force (Rotate One by One)**

Right rotate by 1:

last = nums[-1]

for i in range(len(nums)-1, 0, -1):

    nums[i] = nums[i-1]
    
nums[0] = last

Time → O(n*k) ❌

**⏱ Complexity (Reversal Method)**

*Time → O(n)*

*Space → O(1)*

Interview Best ✔

**🔥 Quick Comparison**

| Rotation Type | Step Order             |
| ------------- | ---------------------- |
| Right Rotate  | Whole → First k → Rest |
| Left Rotate   | First d → Rest → Whole |
