# 2. Palindrome Number
Given an integer x, return true if x is a palindrome, and false otherwise.

**🥇 Approach 1: Convert to String (Simple)**

💡 Idea:

Convert number to string and check if it equals its reverse.

**⏱ Time Complexity:** O(n)

**📦 Space Complexity:** O(n)

```python
def is_palindrome_string(x):
    if x < 0:
        return False
    
    s = str(x)
    return s == s[::-1]

##Driver Code
x = 121
result = is_palindrome_string(x)
print(result)
```

✔ Very easy

✔ Clean

❌ Uses extra space

**🥈 Approach 2: Reverse the Whole Number (Math Based)**

💡 Idea:

Reverse the number mathematically and compare.

**⏱ Time Complexity:** O(log n)

**📦 Space Complexity:** O(1)

```python
def is_palindrome_reverse(x):
    if x < 0:
        return False
    
    original = x
    reversed_num = 0
    
    while x > 0:
        digit = x % 10
        reversed_num = reversed_num * 10 + digit
        x //= 10
    
    return original == reversed_num

##Driver Code
x = 121
result = is_palindrome_reverse(x)
print(result)
```

✔ No string conversion

✔ Interview friendly

**🏆 Approach 3: Reverse Half Only (Most Optimized)**


*🔥 This is what top candidates explain.*

💡 Why Reverse Full Number Is Not Ideal?

If number is very large → possible overflow in some languages (Java/C++).

Better approach:

- Reverse only HALF of the number.

**🧠 Logic**

For palindrome:

First half = reversed second half

Example:

1221

After loop:

x = 12

reversed_half = 12

**⚡ Optimized Code**

```python
def is_palindrome_optimized(x):
    if x < 0 or (x % 10 == 0 and x != 0):
        return False
    
    reversed_half = 0
    
    while x > reversed_half:
        reversed_half = reversed_half * 10 + x % 10
        x //= 10
    
    return x == reversed_half or x == reversed_half // 10

##Driver Code
x = 121
result = is_palindrome_optimized(x)
print(result)
```

🔍 Why This Works?

For odd length:

12321

At end:

x = 12

reversed_half = 123

So we remove middle digit:

reversed_half // 10 = 12

**📊 Comparison**

| Approach     | Time     | Space | Interview Level |
| ------------ | -------- | ----- | --------------- |
| String       | O(n)     | O(n)  | Beginner        |
| Reverse Full | O(log n) | O(1)  | Good            |
| Reverse Half | O(log n) | O(1)  | 🥇 Best         |
