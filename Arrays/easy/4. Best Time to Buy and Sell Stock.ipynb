# 4. Best Time to Buy and Sell Stock

You are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

**🥇 Approach 1: Brute Force**

💡 Idea

*Try every pair (buy, sell).*

**⏱ Time:** O(n²)
**📦 Space:** O(1)

```python
def maxProfit_bruteforce(prices):
    n = len(prices)
    max_profit = 0
    
    for i in range(n):
        for j in range(i + 1, n):
            profit = prices[j] - prices[i]
            max_profit = max(max_profit, profit)
    
    return max_profit

##Driver Code
prices = [7,1,5,3,6,4]
result = maxProfit_bruteforce(prices)
print(result)
```

❌ Too slow for large input.

**🏆 Approach 2: Optimal One-Pass (Greedy)**

This is the correct interview solution.

**🧠 Core Idea**

We want:

Maximum (Sell Price - Minimum Buy Price before it)

So:

- Track minimum price so far
- Calculate profit at each step
- Update max profit

**⏱ Time:** O(n)
**📦 Space:** O(1)

```python
def maxProfit(prices):
    min_price = float ('inf')
    max_profit = 0

    for i in range(len(prices)):
        if prices[i] < min_price:
            min_price = prices[i]
        elif prices[i] - min_price > max_profit:
            max_profit = prices[i] - min_price
    return max_profit

##Driver Code
prices = [7,1,5,3,6,4]
result = maxProfit(prices)
print(result)
```

| Price | Min So Far | Profit | Max Profit |
| ----- | ---------- | ------ | ---------- |
| 7     | 7          | 0      | 0          |
| 1     | 1          | 0      | 0          |
| 5     | 1          | 4      | 4          |
| 3     | 1          | 2      | 4          |
| 6     | 1          | 5      | 5          |
| 4     | 1          | 3      | 5          |


🎯 Why This Works

We are:

- Always buying at lowest seen price
- Always checking best selling opportunity
- This is Greedy Algorithm.

📊 Comparison

| Approach    | Time  | Space | Verdict |
| ----------- | ----- | ----- | ------- |
| Brute Force | O(n²) | O(1)  | ❌      |
| One Pass    | O(n)  | O(1)  | 🥇 Best |


 # Best Time to Buy and Sell Stock II

```python
def maxProfit(prices):
    profit = 0
    
    for i in range(1, len(prices)):
        if prices[i] > prices[i - 1]:
            profit += prices[i] - prices[i - 1]
    
    return profit


# Driver Code
prices = [7,1,5,3,6,4]
print(maxProfit(prices))
```

| i | Compare | Profit Added | Total |

| 3 | 3<5     | 0            | 4     |
| 4 | 6>3     | +3           | 7     |
| 5 | 4<6     | 0            | 7     |
