# 8. Single Number

Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

**🥇 Approach 1: Brute Force**

💡 Idea:

*Count frequency of each number.*

**⏱ Time:** O(n²)
**📦 Space:** O(1)

```python
def singleNumber_bruteforce(nums):
    for num in nums:
        if nums.count(num) == 1:
            return num
            
##Driver Code 
nums = [4,1,2,1,2]
result = singleNumber_bruteforce(nums)
print(result)
```

❌ Very slow

**🥈 Approach 2: Using HashMap**

**⏱ Time:** O(n)

**📦 Space:** O(n)

```python
def singleNumber_hashmap(nums):
    freq = {}
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
    
    for key in freq:
        if freq[key] == 1:
            return key

nums = [4,1,2,1,2]
result = singleNumber_hashmap(nums)
print(result)
```

✔ Good
❌ Extra space

**🏆 Approach 3: XOR Trick (BEST)**

This is the real interview solution.

🧠 Core XOR Properties

a ^ a = 0

a ^ 0 = a

XOR is commutative and associative

So:

2 ^ 2 ^ 1

= (2 ^ 2) ^ 1

= 0 ^ 1

= 1

All duplicate numbers cancel out.

**⏱ Time:** O(n)
**📦 Space:** O(1) ✅ BEST

```python
def singleNumber(nums):
    result = 0
    
    for num in nums:
        result ^= num
    
    return result

nums = [4,1,2,1,2]
result = singleNumber(nums)
print(result)
```

🔍 Dry Run

Input:

[4,1,2,1,2]

Step-by-step:

0 ^ 4 = 4

4 ^ 1 = 5

5 ^ 2 = 7

7 ^ 1 = 6

6 ^ 2 = 4

Answer = 4 ✅
