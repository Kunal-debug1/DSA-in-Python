# 3. Majority Element
Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.

**🧠 Approach 1: Brute Force**

💡 Idea:

*For each element, count its frequency.*

**⏱ Time Complexity:** O(n²)

**📦 Space Complexity:** O(1)

```python
def majority_element_bruteforce(nums):
    n = len(nums)
    
    for i in range(n):
        count = 0
        for j in range(n):
            if nums[j] == nums[i]:
                count += 1
        
        if count > n // 2:
            return nums[i]

##Driver Code
nums = [2,2,1,1,1,2,2]
result = majority_element_bruteforce(nums)
print(result)
```

✔ Simple
❌ Slow

**🚀 Approach 2: Using HashMap (Dictionary)**

💡 Idea:

*Count frequency using dictionary.*

**⏱ Time:** O(n)

**📦 Space:** O(n)

```python
def majority_element_hashmap(nums):
    freq = {}
    n = len(nums)
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
        
        if freq[num] > n // 2:
            return num

##Driver Code
nums = [2,2,1,1,1,2,2]
result = majority_element_hashmap(nums)
print(result)
```

✔ Efficient
✔ Easy to understand
✔ Common interview solution

**🏆 Approach 3: Boyer-Moore Voting Algorithm (🔥 BEST)**

This is the real interview winner.

**💡 Core Idea**

If majority element appears more than n/2 times,

it will survive elimination process.

We:

- Keep a candidate
- Increase count if same
- Decrease count if different
- If count becomes 0 → change candidate

**⏱ Time:** O(n)

**📦 Space:** O(1) ✅ BEST

```python
def majority_element(nums):
    count = 0
    candidate = None
    
    for num in nums:
        if count == 0:
            candidate = num
        
        if num == candidate:
            count += 1
        else:
            count -= 1
    
    return candidate

##Driver Code
nums = [2,2,1,1,1,2,2]
result = majority_element(nums)
print(result)
```

| --- | --------- | ----- |

| 2   | 2         | 2     |
| 1   | 2         | 1     |
| 1   | 2         | 0     |
| 1   | 1         | 1     |
| 2   | 1         | 0     |
| 2   | 2         | 1     |
