# 9. Single Number

Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

**🔥 Optimal Approach → XOR Trick (Very Important)**

*🧠 Key Properties of XOR:*

a ^ a = 0

a ^ 0 = a

XOR is commutative and associative

So:

2 ^ 2 ^ 1

= (2 ^ 2) ^ 1

= 0 ^ 1

= 1

Duplicates cancel out 🔥

✅ Optimal Code (O(n), O(1))

```python
def singleNumber(nums):
    result = 0
    
    for num in nums:
        result ^= num
    
    return result


# Driver Code
nums = [4,1,2,1,2]
print(singleNumber(nums))
```

🔎 Dry Run

[4,1,2,1,2]

Step-by-step:

0 ^ 4 = 4

4 ^ 1 = 5

5 ^ 2 = 7

7 ^ 1 = 6

6 ^ 2 = 4

✔ Answer = 4

**⏱ Complexity**

Time → O(n)

Space → O(1)

Best possible ✔

**❌ Brute Force (Not Recommended)**

```python
def singleNumber(nums):
    for num in nums:
        if nums.count(num) == 1:
            return num

nums = [4,1,2,1,2]
print(singleNumber(nums))
```

Time → O(n²) ❌

**❌ Using HashSet**

```python
def singleNumber(nums):
    seen = set()
    
    for num in nums:
        if num in seen:
            seen.remove(num)
        else:
            seen.add(num)
    
    return seen.pop()

nums = [4,1,2,1,2]
print(singleNumber(nums))
```

Time → O(n)

Space → O(n) ❌

XOR is better.
