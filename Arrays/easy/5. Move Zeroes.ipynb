# 5. Move Zeroes 
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

**🥇 Approach 1: Brute Force (Extra Array)**

💡 Idea:

- Store non-zero elements
- Count zeros
- Append zeros at end

**⏱ Time:** O(n)
**📦 Space:** O(n)

```python
def moveZeroes_bruteforce(nums):
    non_zero = []
    zero_count = 0
    
    for num in nums:
        if num != 0:
            non_zero.append(num)
        else:
            zero_count += 1
    
    return non_zero + [0] * zero_count

# Driver Code
nums = [0,1,0,3,12]
result = moveZeroes_bruteforce(nums)
print(result)
```

❌ Not allowed in interview (uses extra space).

**🥈 Approach 2: Two Pointer (Best – In Place)**

💡 Core Idea

- Use pointer i to track position for next non-zero element.
- Traverse with j
- If nums[j] != 0 → swap with nums[i]
- Increment i

**⏱ Time:** O(n)
**📦 Space:** O(1) ✅

```python
def moveZeroes(nums):
    i = 0
    
    for j in range(len(nums)):
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1


# Driver Code
nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```

| j | nums[j] | i | Action                   |
| - | ------- | - | ------------------------ |
| 0 | 0       | 0 | Skip                     |
| 1 | 1       | 0 | Swap(0,1) → [1,0,0,3,12] |
| 2 | 0       | 1 | Skip                     |
| 3 | 3       | 1 | Swap(1,3) → [1,3,0,0,12] |
| 4 | 12      | 2 | Swap(2,4) → [1,3,12,0,0] |


🏆 Why This Is Best?

- Single pass
- No extra space
- Maintains order
- Very common interview pattern
- Pattern = Two Pointers

**🎯 Alternative Optimized Version (Less Swaps)**

Instead of swapping every time, first move non-zero forward, then fill zeros.

```python
def moveZeroes(nums):
    i = 0
    
    # Move non-zero elements
    for num in nums:
        if num != 0:
            nums[i] = num
            i += 1
    
    # Fill remaining with zeros
    while i < len(nums):
        nums[i] = 0
        i += 1

# Driver Code
nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```

✔ Same complexity
✔ Fewer swaps

📊 Comparison

| Approach    | Time | Space | Interview |
| ----------- | ---- | ----- | --------- |
| Extra Array | O(n) | O(n)  | ❌        |
| Two Pointer | O(n) | O(1)  | 🥇 Best   |
