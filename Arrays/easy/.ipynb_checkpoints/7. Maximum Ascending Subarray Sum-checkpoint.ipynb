# 7. Maximum Ascending Subarray Sum

Given an array of positive integers nums, return the maximum possible sum of an strictly increasing subarray in nums.

A subarray is defined as a contiguous sequence of numbers in an array.

🧠 Core Idea

We just need:

- Track current ascending sum
- Reset when order breaks
- Keep updating max sum

*This is similar to Kadane’s Algorithm, but condition-based.*

**🥇 Optimal Solution (One Pass)**

**⏱ Time:** O(n)

**📦 Space:** O(1)

```python
def maxAscendingSum(nums):
    max_sum = nums[0]
    current_sum = nums[0]
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_sum += nums[i]
        else:
            current_sum = nums[i]
        
        max_sum = max(max_sum, current_sum)
    
    return max_sum

##Driver code
nums = [10,20,30,5,10,50]
ans = maxAscendingSum(nums)
print(ans)
```

🔍 Dry Run

Input:

[10,20,30,5,10,50]

❗ Edge Cases

1️⃣ All decreasing

[5,4,3,2]

Answer = 5

2️⃣ Single element

[7]

Answer = 7
