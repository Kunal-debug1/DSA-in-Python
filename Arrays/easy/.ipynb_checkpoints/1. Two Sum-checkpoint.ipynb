# 1. Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have **exactly one solution**, and you may not use the same element twice.

You can return the answer in any order.

**Example 1:**

**🧠 Approach 1: Brute Force (Naive Method)**

💡 Idea:
- Check every pair of elements.

⏱ Time Complexity:            O(n²)

📦 Space Complexity:           O(1)

```python
def two_sum_bruteforce(nums, target):
    n = len(nums)
    
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]


##Driver Code 
nums =  [2,7,11,15]
target = 9
result = two_sum_bruteforce(nums, target)
print(result)
```

Why it works:

- Try all combinations
- Return when match found

**🚀 Approach 2: Better Approach (Using HashMap / Dictionary)**

💡 Idea:

Store numbers in dictionary while iterating.
Check if complement exists.

*Complement = target - current_number*

**⏱ Time Complexity:** O(n)

**📦 Space Complexity:** O(n)

```python
def two_sum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
            if target - num in seen:
                return [seen[target - num], i]

            seen[num] = i

##Driver Code 
nums =  [2,7,11,15]
target = 9
result = two_sum(nums, target)
print(result)
```

**⚡ Approach 3: Using Sorting + Two Pointers (Advanced Thinking)**

💡 Idea:

- Sort array
- Use two pointers
- Move pointers based on sum

⚠️ Problem:

- Sorting changes original indices.

- So we store (value, index) pair.

**⏱ Time Complexity:**  O(n log n)

**📦 Space Complexity:** O(n)

```python
def two_sum_two_pointers(nums, target):
    arr = [(nums[i], i) for i in range(len(nums))]
    arr.sort()
    
    left = 0
    right = len(arr) - 1
    
    while left < right:
        current_sum = arr[left][0] + arr[right][0]
        
        if current_sum == target:
            return [arr[left][1], arr[right][1]]
        
        elif current_sum < target:
            left += 1
        else:
            right -= 1

##Driver Code 
nums =  [2,7,11,15]
target = 9
result = two_sum_two_pointers(nums, target)
print(result)
```

**🏆 Which Approach is BEST?**

For the Two Sum (return values version):

**🥇 Best: HashMap / Set Approach**

*🔍 Why Is This the Best?*

1️⃣ Time Complexity

- Brute Force → O(n²)
- Sorting + Two Pointers → O(n log n)
- HashMap → O(n) ✅
- Linear time is the fastest possible here.


2️⃣ No Sorting Needed

Sorting:

- Changes order
- Adds extra time complexity
- Complicates index tracking

HashMap:

- Works in one pass
- No modification of array
- Clean logic

3️⃣ Interview Perspective

If you:

- Explain brute force first
- Then optimize to HashMap
- Then explain complexity improvement
- Interviewer will be impressed.

Because you showed:

- Problem-solving
- Optimization thinking
- Complexity awareness
