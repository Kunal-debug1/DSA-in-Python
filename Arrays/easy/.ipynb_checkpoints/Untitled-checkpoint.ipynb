```python
# Sum of array
arr = [1, 2, 3, 4]

total = 0

for num in arr:
    total += num

print(total)
```

```python
# Mean of array
def findMean(arr):
        # code here 
        n = len(arr)
        sum = 0
        
        for num in arr:
            sum += num
            
        return sum//n
    
arr = [1, 3, 4, 2, 6, 5, 8, 7]        
print(findMean(arr))
```

```python
# Remove Duplicates
def removeDuplicates(arr):
    if not arr:
        return arr
    
    i = 0
    
    for j in range(1, len(arr)):
        if arr[j] != arr[i]:
            i += 1
            arr[i] = arr[j]
    
    return arr[:i+1]


arr = [1,2,2,3,4,4,5]
print(removeDuplicates(arr))
```

```python
# array is sorted or not
def isSorted(arr) -> bool:
        # code here
        n = len(arr)
        
        for i in range(1, n):
            if arr[i] < arr[i - 1]:
                return False
             
        return True
    
arr = [10, 20, 30, 40, 50]
print(isSorted(arr))
```

```python
#Given an array arr[] consisting of only 0’s and 1’s, return count of the maximum number of consecutive 1’s or 0’s present in the array. 
def maxConsecBits(arr):
        #code here 
        ans = 1
        count = 1
        
        for i in range(1,len(arr)):
            if arr[i] == arr[i - 1]:
                count += 1
            else:
                count = 1
            
            ans = max(ans, count)
            
        return ans
arr = [0, 1, 0, 1, 1, 1, 1]
print(maxConsecBits(arr))
```
