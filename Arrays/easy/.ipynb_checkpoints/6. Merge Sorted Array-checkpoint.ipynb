# 6. Merge Sorted Array
You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

**🥇 Optimal Approach (Three Pointers – Best)**

💡 Idea

Use:

- i = m - 1 → last valid element in nums1

- j = n - 1 → last element in nums2

- k = m + n - 1 → last position of nums1

- Compare from back and fill largest first.

**⏱ Time:** O(m + n)
**📦 Space:** O(1) ✅

```python
def merge(nums1, m, nums2, n):
    i = m - 1
    j = n - 1
    k = m + n - 1
    
    while i >= 0 and j >= 0:
        if nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1
        k -= 1
    
    while j >= 0:
        nums1[k] = nums2[j]
        j -= 1
        k -= 1


# Driver Code
nums1 = [1,2,3,0,0,0]
nums2 = [2,5,6]
m = 3
n = 3

merge(nums1, m, nums2, n)
print(nums1)
```

🔍 Dry Run

Input:

nums1 = [1,2,3,0,0,0]

nums2 = [2,5,6]

Step-by-step:

| i | j | k | nums1         |
| - | - | - | ------------- |
| 2 | 2 | 5 | [1,2,3,0,0,6] |
| 2 | 1 | 4 | [1,2,3,0,5,6] |
| 2 | 0 | 3 | [1,2,3,3,5,6] |
| 1 | 0 | 2 | [1,2,2,3,5,6] |
| 0 | - | - | Done          |


Final:

[1,2,2,3,5,6]

❗ Why We Only Check Remaining nums2?

Because if nums1 still has elements → they’re already in correct place.

**🥈 Brute Force (Not Interview Friendly)**

```python
nums1[:m] = sorted(nums1[:m] + nums2)
```
